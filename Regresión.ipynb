## Función hipótesis
$$
h_{\theta}(x)=\theta_0+\theta_1x
$$

## Construcción del modelo

Buscar los parámetros $\theta_0$ y $\theta_1$ que generen la función hipótesis ($h_{\theta}(x)$) que mejor se adapte al conjunto de datos de entrenamiento $(x,y)$

Se minimiza una función de costo ($J(\theta)$) para obtener los parametros $\theta_0$ y $\theta_1$ óptimos

## Función de costo

Error cuadrático medio (mean square error (MSE))
$$
J(\theta)=\frac{1}{2n}\sum_{i=1}^n(h_{\theta}(x_i)-y_i)^2=MSE(\theta)
$$

<img src="fig4.png" alt="Drawing" style="width: 500px;"/>

## Función de optimización (Gradient Descent)
$$
\min_{\theta}J(\theta)
$$

El algoritmo iteractivo consiste en repetir el siguiente proceso hasta que $\theta_0$ y $\theta_1$ no varien
\begin{eqnarray*}
\hat{\theta_0}&=\theta_0-\alpha\ \frac{\delta J(\theta)}{\delta\theta_0}=\theta_0-\alpha\frac{1}{n}\sum_{i=1}^n(h_{\theta}(x_i)-y_i) \\
\hat{\theta_1}&=\theta_1-\alpha\ \frac{\delta J(\theta)}{\delta\theta_1}=\theta_1-\alpha\frac{1}{n}\sum_{i=1}^n x_i(h_{\theta}(x_i)-y_i) \\
\end{eqnarray*}
donde $\alpha$ es la tasa de aprendizaje (**learning rate**), que determina el tamaño del paso en cada iteración

# Regularización

## Regresión Ridge

La regresión Ridge es una técnica de regularización utilizada en la regresión lineal para evitar el sobreajuste. Funciona agregando un término de penalización a la función de costo que minimiza la suma de los errores al cuadrado. El término de penalización es una función de los coeficientes del modelo de regresión, y reduce sus valores hacia cero.

La función de costo de regresión Ridge se puede escribir como:
$$
J(\theta) = MSE(\theta) + \alpha\sum_{i=1}^n\theta_i^2
$$
donde $J(\theta)$ es la función de costo, MSE($\theta$) es el error cuadrático medio entre los valores predichos y los valores reales, $\theta$ es el vector de coeficientes de regresión, $\alpha$ es el parámetro de **regularización** y $\sum_{i=1}^n\theta_i^2$ es la suma de cuadrados de los coeficientes, **A esta penalización se le conoce como** $\mathcal{l}_2$.

El parámetro de regularización $\alpha$ controla el grado de regularización. Un valor mayor de $\alpha$ conduce a una regularización más fuerte, lo que significa que los coeficientes se reducen más hacia cero. Por otro lado, un valor menor de $\alpha$ conduce a una regularización más débil, lo que significa que los coeficientes están menos restringidos.

La regresión Ridge es útil cuando se trata de multicolinealidad, que es una situación en la que las variables independientes están altamente correlacionadas entre sí. En tales casos, el método de los mínimos cuadrados ordinarios puede producir estimaciones inestables y poco fiables de los coeficientes. La regresión de crestas puede ayudar a estabilizar las estimaciones y mejorar el rendimiento predictivo del modelo.

## Regresión Lasso

La regresión Lasso es otra técnica de regularización utilizada en la regresión lineal para evitar el sobreajuste. Al igual que la regresión Ridge, agrega un término de penalización a la función de costo para controlar el tamaño de los coeficientes, pero el término de penalización utilizado en la regresión Lasso es diferente.

La función de costo de la regresión lasso se puede escribir como:
$$
J(\theta) = MSE(\theta) + \alpha\sum_{i=1}^n|\theta_i|
$$
donde $J(\theta)$ es la función de costo, MSE($\theta$) es el error cuadrático medio entre los valores predichos y los valores reales, $\theta$ es el vector de coeficientes de regresión, $\alpha$ es el parámetro de regularización y $\sum_{i=1}^n|\theta_i|$ es la suma de los valores absolutos de los coeficientes, **A esta penalización se le conoce como** $\mathcal{l}_1$.

A diferencia de la regresión Ridge, la regresión Lasso puede reducir los coeficientes a cero, lo que significa que puede realizar una selección de características y excluir del modelo los predictores menos importantes. Esta propiedad de la regresión Lasso la hace particularmente útil cuando se trata de datos de alta dimensión donde el número de predictores es mucho mayor que el número de observaciones.

La elección del parámetro de regularización $\alpha$ controla la cantidad de regularización en el modelo. Un valor mayor de $\alpha$ conduce a una regularización más fuerte, lo que significa que se reducen más coeficientes hacia cero. Por otro lado, un valor menor de $\alpha$ conduce a una regularización más débil, lo que significa que se excluyen menos coeficientes del modelo.

En general, la regresión Lasso es una herramienta poderosa tanto para la regularización como para la selección de características en modelos de regresión lineal, y puede mejorar la interpretabilidad y el rendimiento predictivo del modelo.

## Elastic net

La técnica elastic net de regularización híbrida, combina la regresión ridge y lasso. Agrega penalizaciones de regularización $\mathcal{l}_1$ y $\mathcal{l}_2$ a la función de costo para controlar el tamaño de los coeficientes y realizar la selección de características.

La función de costo elastic net se puede escribir como:
$$
J(\theta) = MSE(\theta) + \alpha\sum_{i=1}^n(\rho|\theta_i|+\frac{1}{2}(1-\rho)\theta_i^2)
$$
$\lambda$ es el parámetro de regularización, $\rho$ es el parámetro de mezcla que controla el equilibrio entre la regularización $\mathcal{l}_1$ y $\mathcal{l}_2$, y $\sum(\rho|\theta_i|+\frac{1}{2}(1-\rho)\theta_i^2)$ es la suma de las penalizaciones $\mathcal{l}_1$ y $\mathcal{l}_2$. El grado en que influye cada una de las penalizaciones está controlado por el hiperparámetro $\rho\in [0,1]$. Cuando $\rho=0$, se ajusta una ridge y cuando $\rho=1$ se ajusta una lasso.

La penalización $\mathcal{l}_1\ (\sum\rho|\theta_i|)$ fomenta la escasez al reducir algunos coeficientes exactamente a cero, mientras que la penalización $\mathcal{l}_2\ (\sum\frac{1}{2}(1-\rho)\theta_i^2)$ fomenta la suavidad al reducir los coeficientes restantes a cero. Una estrategia frecuentemente utilizada es asignarle casi todo el peso a la penalización $\mathcal{l}_1$ ($\rho$) muy próximo a 1) para conseguir seleccionar predictores y un poco a la $\mathcal{l}_2$ para dar cierta estabilidad en el caso de que algunos predictores estén correlacionados.

Elastic net es particularmente útil cuando se trata de datos de gran dimensión en los que se necesitan tanto la regularización como la selección de características. Puede mejorar la interpretabilidad y el rendimiento predictivo del modelo seleccionando un subconjunto de los predictores más importantes y reduciendo el impacto del ruido y la multicolinealidad.

En general, elastic net es una técnica de regularización flexible y poderosa que combina las fortalezas de la regresión ridge y lasso y puede manejar una amplia gama de problemas de regresión lineal.

## Métricas de Evaluación

Para llevar un seguimiento de si nuestro modelo se comporta correctamente, es preciso aplicar métricas de evaluación. Esta métrica de evaluación es el error calculado entre la recta generada (o el hiperplano) a los puntos reales y será la función que hay qque minimizar usando el descenso del gradiente.

Algunas de las más comunes en regresión son:

- **Error Absoluto Medio (MAE):**
$$
MAE(\theta)=\frac{1}{n}\sum_{i=1}^n |h_{\theta}(x_i)-y_i|
$$

EL error absoluto medio o “MAE”, es la media de la diferencia absoluta entre los puntos de datos reales, y la salida predicha. Si tomamos esto como la estrategia a seguir, a cada paso del exceso de gradiente, se reducirá el MAE.

- **Error cuadrático medio (MSE):**
$$
MSE(\theta)=\frac{1}{2n}\sum_{i=1}^n(h_{\theta}(x_i)-y_i)^2
$$

El error cuadrático medio o “MSE”, es la media de la diferencia entre los puntos reales de datos y la salida predicha, al cuadrado. Este método penaliza más las diferencias mayores y es la estándar en los problemas de regresión.

Si elegimos esta estrategia, cada paso del descenso de gradiente reducirá el MSE. Esto hace que sea el método preferido para calcular la recta que mejor se ajuste , y es también llamado “ Ordinary Least Squares” o OLS.

- **Raiz del Error Cuadrático Medio (RMSE):**

El RMSE por sus siglas en inglés, es la raíz de la media de los errores elevados al cuadrado. Éste es el medidor de evaluación mas popular para determinar el desempeño de los modelos de regresión ya que la raíz, consigue las mismas unidades que la y.
$$
RMSE=\sqrt{\frac{1}{2n}\sum_{i=1}^n(h_{\theta}(x_i)-y_i)^2}
$$

- **Coeficiente de determinación o $R^2$:**

El coeficiente de determinación se puede entender como una versión estandarizada del MSE, que proporciona una mejor interpretación del rendimiento del modelo.

Técnicamente, el R² representa la varianza de las respuesta capturada por el modelo:
$$
R^2=1-\frac{SSE}{SST},
$$
donde

\begin{align}
SST&=\sum_{i=1}^n(y_i-\bar{y})^2\\
SSE&=\sum_{i=1}^n(h_{\theta}(x_i)-y_i)^2
\end{align}

- **$R^2$ ajustado**

Al igual que $R^2$, el $R^2$ ajustado también muestra qué tan bien se ajustan los términos a una curva o línea, pero se ajusta al número de términos de un modelo. Se da en la siguiente fórmula:
$$
R_{adj}^2=1-\left[\frac{(1-R^2)(n-1)}{n-k-1}\right]
$$
donde $n$ es el número de observaciones y $k$ es el numero de predictores.

### Tolerancia al riesgo de los inversores

Para predecir la tolerancia al riesgo de un individuo, utilizaremos los datos de la Encuesta de Finanzas del Consumidor (SCF), realizada por la Junta de la Reserva Federal. La encuesta incluye respuestas sobre la demografía de los hogares, el patrimonio neto y los activos financieros y no financieros para el mismo conjunto de personas en 2007 (antes de la crisis) y 2009 (poscrisis). Esto nos permite ver cómo cambió la asignación de cada hogar después de la crisis financiera mundial de 2008. La fuente de los datos: <https://www.federalreserve.gov/econres/scf_2009p.htm>

Las librerías utilizadas en este ejemplo son:

```python
# Tratamiento de datos
# ==============================================================================
import pandas as pd
import numpy as np
import copy

# Gráficos
# ==============================================================================
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns


# Preprocesado y modelado
# ==============================================================================
from sklearn.preprocessing import StandardScaler
from scipy.stats import pearsonr
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet

# Configuración warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

## Cargamos los datos

```python
dataset = pd.read_excel('SCFP2009panel.xlsx')
```

## Preparación de datos y selección de funciones

## Preparamos la variable de predicción

El conjunto de datos de la "Encuesta de finanzas del consumidor" contiene la demografía, el patrimonio neto y los activos financieros y no financieros del hogar para la misma demografía en 2007 (antes de la crisis) y 2009 (poscrisis).

Preparamos la variable de predicción, que es la "verdadera" tolerancia al riesgo en los siguientes pasos. Hay diferentes maneras de obtener la "verdadera" tolerancia al riesgo. La idea y el propósito de este estudio es encontrar un enfoque para resolver el problema de las finanzas conductuales mediante el aprendizaje automático.

Los pasos para calcular las variables de predicción son los siguientes:

1. Calcule el activo riesgoso y los activos sin riesgo para todos los individuos en los datos de la encuesta. Los activos riesgosos y sin riesgo se definen de la siguiente manera:

- **Los activos de riesgo** son inversiones en fondos mutuos, acciones, bonos, materias primas y bienes raíces, y una estimación del capital humano.

- **Activos Libres de Riesgo:** saldos de cuentas corrientes y de ahorro, certificados de depósito y otros saldos de efectivo y equivalentes.

2. Tomamos la relación entre los activos de riesgo y los activos totales de un inversor y la consideramos como una medida de la tolerancia al riesgo de un inversor. A partir de los datos de SCF, tenemos los datos de activos riesgosos y sin riesgo para los individuos para 2007 y 2009. Usamos estos datos y normalizamos los activos riesgosos con el precio de las acciones de 2007 frente a 2009 para obtener la tolerancia al riesgo.

- **La tolerancia al riesgo** se define simplemente como la relación entre activos riesgosos y activos sin riesgo normalizada con el S&P500 promedio de 2007 frente a 2009. S&P500 promedio en 2007: 1478 S&P500 promedio en 2009: 948

3. En mucha literatura, un inversor inteligente es aquel que no cambia su tolerancia al riesgo durante los cambios en el mercado. Así, consideramos inversores inteligentes a los inversores que cambian su tolerancia al riesgo en menos del 10% entre 2007 y 2009. Por supuesto, este es un juicio cualitativo y está sujeto a cambios. Sin embargo, como se mencionó anteriormente, más que ser exacto y preciso, el propósito de este estudio de caso es demostrar el uso del aprendizaje automático y proporcionar un marco basado en el aprendizaje automático en las finanzas conductuales y la gestión de carteras que se puede aprovechar aún más para un análisis más detallado.

```python
# Promedio SP500 durante el 2007 y 2009
Average_SP500_2007=1478
Average_SP500_2009=948

# Tolerancia al riesgo en 2007
dataset['RiskFree07']= dataset['LIQ07'] + dataset['CDS07'] + dataset['SAVBND07'] + dataset['CASHLI07']
dataset['Risky07'] = dataset['NMMF07'] + dataset['STOCKS07'] + dataset['BOND07'] 
dataset['RT07'] = dataset['Risky07']/(dataset['Risky07']+dataset['RiskFree07'])

# Tolerancia al riesgo en 2009
dataset['RiskFree09']= dataset['LIQ09'] + dataset['CDS09'] + dataset['SAVBND09'] + dataset['CASHLI09']
dataset['Risky09'] = dataset['NMMF09'] + dataset['STOCKS09'] + dataset['BOND09'] 
dataset['RT09'] = dataset['Risky09']/(dataset['Risky09']+dataset['RiskFree09'])*\
                (Average_SP500_2009/Average_SP500_2007)
```

Veamos los detalles de los datos:

```python
dataset.head()
```

Calculemos el cambio porcentual en la tolerancia al riesgo entre 2007 y 2009.

```python
dataset2 = copy.deepcopy(dataset)
dataset2['PercentageChange'] = np.abs(dataset2['RT09']/dataset2['RT07']-1)
```

Comprobando las filas con valores nulos o nan y eliminándolas.

```python
dataset2=dataset2.dropna(axis=0)
dataset2=dataset2[~dataset2.isin([np.nan, np.inf, -np.inf]).any(1)]
```

```python
dataset2.shape
```

Investiguemos el comportamiento de tolerancia al riesgo de los individuos en 2007 versus 2009. Primero miramos la tolerancia al riesgo en 2007:

```python
sns.distplot(dataset2['RT07'], 
             hist=True, 
             kde=False,
             bins=int(180/5), 
             color = 'blue',
             hist_kws={'edgecolor':'black'})
plt.show()
```

Si observamos la tolerancia al riesgo de 2007, vemos que un número significativo de personas tenían una tolerancia al riesgo cercana a uno, lo que significa que la inversión estaba más sesgada hacia los activos de riesgo en comparación con los activos sin riesgo. Veamos ahora la tolerancia al riesgo de 2009.

```python
sns.distplot(dataset2['RT09'], 
             hist=True, 
             kde=False,
             bins=int(180/5), 
             color = 'blue',
             hist_kws={'edgecolor':'black'})
plt.show()
```

Es evidente que el comportamiento de los individuos se revirtió después de la crisis. La tolerancia general al riesgo disminuyó, como lo demuestra la enorme proporción de hogares que tenían una tolerancia al riesgo cercana a cero en 2009. La mayoría de las inversiones de estas personas se realizaron en activos libres de riesgo.

En el siguiente paso, elegimos a los inversores inteligentes cuyo cambio en la tolerancia al riesgo entre 2007 y 2009 fue inferior al 10%. El propósito sería predecir la verdadera tolerancia al riesgo de un individuo dadas las características demográficas, financieras y de disposición a asumir riesgos.

```python
dataset3 = copy.deepcopy(dataset2)  
dataset3 = dataset3[dataset3['PercentageChange']<=.1]
dataset3['TrueRiskTolerance'] = (dataset3['RT07'] + dataset3['RT09'])/2
```

Eliminemos otras etiquetas que podrían no ser necesarias para la predicción.

```python
dataset3.drop(labels=['RT07', 'RT09'], axis=1, inplace=True)
dataset3.drop(labels=['PercentageChange'], axis=1, inplace=True)
```

# Evaluar algoritmos y modelos

Evaluemos los algoritmos y los modelos.

## División train y test

```python
# dividir el conjunto de datos de validación para el final
Y= dataset3["TrueRiskTolerance"]
X = dataset3.loc[:, dataset3.columns != 'TrueRiskTolerance']
# scaler = StandardScaler().fit(X)
# rescaledX = scaler.transform(X)
validation_size = 0.2
seed = 3
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

## Modelos de regresión OLS

```python
# Modelo de regresión
# ==============================================================================
LR = LinearRegression()
LR.fit(X = X_train, y = Y_train)
```

```python
# Coeficientes del modelo
# ==============================================================================
df_coeficientes = pd.DataFrame(
                        {'predictor': X_train.columns,
                         'coef': LR.coef_.flatten()}
                  )

# Predictores incluidos en el modelo (coeficiente != 0)
df_coeficientes[df_coeficientes.coef != 0]
```

```python
# Predicciones de validación
# ==============================================================================
predicciones = LR.predict(X=X_test)
predicciones = predicciones.flatten()
predicciones[:10]
```

```python
# Error de test del modelo 
# ==============================================================================
rmse_ols = mean_squared_error(
            y_true  = Y_test,
            y_pred  = predicciones,
            squared = False
           )
print("")
print(f"El error (rmse) de test es: {rmse_ols}")
```

## Modelo de regresión Ridge

```python
# Modelo de regresión Ridge
# ==============================================================================
RIDGE = Ridge()
RIDGE.fit(X = X_train, y = Y_train)
```

```python
# Coeficientes del modelo
# ==============================================================================
df_coeficientes = pd.DataFrame(
                        {'predictor': X_train.columns,
                         'coef': RIDGE.coef_.flatten()}
                  )

# Predictores incluidos en el modelo (coeficiente != 0)
df_coeficientes[df_coeficientes.coef != 0]
```

```python
# Error de test del modelo 
# ==============================================================================
rmse_Ridge = mean_squared_error(
            y_true  = Y_test,
            y_pred  = predicciones,
            squared = False
           )
print("")
print(f"El error (rmse) de test es: {rmse_Ridge}")
```
